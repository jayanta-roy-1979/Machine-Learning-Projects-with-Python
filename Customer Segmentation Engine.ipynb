```python
# We want to perform clusters of customers who share similar bebaviour for that lets select that columns 
# Spending Score and Annual Income

# lets that Import the warnings Library
import warnings
warnings.filterwarnings('ignore')

# Lets Select the Spending Score and Annual Income from the data
x=data.loc[:,['Spending Score (1-100)','Annual Income (k$)']].values

# Lets check the Shape of x
print(x.shape)
```

```python
# Lets also check the data, which we are going to use for the Clustering Analysis 
x_data=pd.DataFrame(x)
x_data.head()
```

```python
#Using Dendrograms to find the no. of Optimal Clusters
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogam created by Jayanta Roy', fontsize = 20)
plt.xlabel('Customers')
plt.ylabel('Ecuclidean Distance')
plt.show()
```

```python
# To Elbow Method to find the No. of Optimal Clusters

from sklearn.cluster import KMeans
#data = list(zip(x, y))
inertias = []
#wcss=[]

for i in range(1,11):
    kmeans= KMeans(n_clusters=i,init='k-means++',max_iter=300, n_init=10,random_state=0)
    kmeans.fit(x.reshape(-1, 1))
    #wcss.append(km.inertia_)
    inertias.append(kmeans.inertia_)
    
plt.plot(range(1,11),inertias, marker='o', ms = 20,mec = 'darkblue', mfc = 'darkblue')
plt.title('The Elbow Method created by Jayanta Roy',fontsize=20)
plt.xlabel('No. of Clusters')
plt.ylabel('Inertia')
plt.show()
    
```

```python
# lets visualize this Clusters
plt.style.use('fivethirtyeight')

km= KMeans(n_clusters=i, init='k-means++',max_iter=300, n_init=10,random_state=0)
y_means=kmeans.fit_predict(x)

plt.scatter(x[y_means==0,0],x[y_means==0,1],s=100,c='steelblue',label='miser')
plt.scatter(x[y_means==1,0],x[y_means==1,1],s=100,c='red',label='general')
plt.scatter(x[y_means==2,0],x[y_means==2,1],s=100,c='green',label='target')
plt.scatter(x[y_means==3,0],x[y_means==3,1],s=100,c='maroon',label='spendthrift')
plt.scatter(x[y_means==4,0],x[y_means==4,1],s=100,c='blue',label='careful')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],50,c='orange',label='careful')

plt.style.use('fivethirtyeight')
plt.title('K Means Clustering between Annual Income and Spending Score created by Jayanta Roy ', fontsize=20)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()
plt.show()
```

```python
# Clustering between Age and Spending Score
from sklearn.cluster import KMeans
#data = list(zip(x, y))
inertias = []
#wcss=[]

for i in range(1,11):
    kmeans= KMeans(n_clusters=i,init='k-means++',max_iter=300, n_init=10,random_state=0)
    kmeans.fit(x)
    #wcss.append(km.inertia_)
    inertias.append(kmeans.inertia_)
    
plt.rcParams['figure.figsize']  =(15,5)
plt.plot(range(1,11),inertias, marker ='o', mec = 'red', mfc = 'red',ms=20)
plt.title('K Means Clustering (The Elbow Method) created by Jayanta Roy',fontsize=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid()
plt.show()
```
